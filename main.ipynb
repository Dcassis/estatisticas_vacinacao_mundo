```python
media_pais['daily_vaccinations'] = media_pais['daily_vaccinations'].astype(int)
media_pais
```

```python
media_pais = media_pais.rename(columns={'country':'pais', 'daily_vaccinations': 'media_vacinacao'})
media_pais
```
